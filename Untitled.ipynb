```python
from linearmodels.iv import IV2SLS
import numpy as np 

# --- データフレーム 'df' は前回の推定で使用した架空データを使用 ---

# 1. 変数の定義
dependent = df['N_C'] # 被説明変数: C
exog = pd.DataFrame(1.0, index=df.index, columns=['const']) # 外生変数: 定数項
df["Yd"] = df['N_GDP'] - df['N_TAX'] - df['N_TR']
endog = df['Yd'] # 内生変数: Yd (可処分所得)

instruments = df[['N_G', 'i']] # 操作変数: G, i

# 2. モデルの定義と推定
# IV2SLS(dependent, exog, endog, instruments)
iv_model = IV2SLS(dependent, exog, endog, instruments)
iv_res = iv_model.fit(cov_type='robust') 

# 3. 結果の確認（係数を取り出す）
# C = c0 + c1 * Yd
c0 = iv_res.params['const'] # 定数項 c0
c1 = iv_res.params['Yd']    # 限界消費性向 c1

print("--- Linearmodelsによる消費関数の推定結果 ---")
print(iv_res)
```

```python
# ブラウン運動かつ単位根過程かつランダムウォーク
# ブラウン運動は微小な連続変化なので厳密には異なるが式は同じもの
# 単位根過程はARモデルの係数が1のもの、二次の係数でも1のときに単位根過程という
# ランダムウォークは単位根過程の中でも、一次のものを指す
# 今回はposition += stepがポイント
# xt = xt-1 + e
import random

position = 0
walk = [position]
for i in range(100):
    step = random.gauss(0, 1)
    position += step
    walk.append(position)

plt.plot(walk)
```

```python
def AR(x):
        if (x >= 1) | (x <= -1):
            print("error")
        else:
            position = 0
            walk = [position]
            for i in range(100):
                e = random.gauss(0, 1)
                position = position * x + e
                walk.append(position)
        plt.plot(walk)
AR(-0.7)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# ----------------------------------------------------
# データセット df が既にロードされていると仮定
# ----------------------------------------------------

# 1. 名目変数（N_I, N_GDPなど）の対数差分化
# N_I (投資) の変化率
df['dln_I'] = np.log(df['N_I']).diff()

# N_GDP (景気指標) の変化率
df['dln_GDP'] = np.log(df['N_GDP']).diff()
# 2. 金利 (r) の差分化
# 金利はすでに率なので、差分を取ります (政策変更のショックとして扱う)
df['d_r'] = df['i'].diff() 

# 3. 必要な説明変数の作成 (投資関数 I_t のベース)
# I_t = i0 + i1*Y_t + i2*r_t + ... を変化率モデルに変換
# 景気 (dln_GDP) と 金利 (d_r) を基本の説明変数とします。

# 変換後のデータは、最初の行がNaNになります。

def I_aic(n, df):
    """
    投資関数モデルにラグ次数 n を含めて推定し、AICを計算します。
    被説明変数: dln_I (投資変化率)
    説明変数: dln_GDP (景気変化率), d_r (金利差分), dln_I のラグ k=1..n
    """
    
    # 1. 被説明変数 (y) と基本の説明変数 (X_base) の定義
    y = df["dln_I"]
    # 投資関数に必要な基本の説明変数は dln_GDP と d_r の2つです
    X_base = df[["dln_GDP", "d_r"]]
    
    # 2. ラグ変数の作成と結合
    X_lag_list = []
    
    # ラグを n=0 から開始すると処理が複雑になるため、n=1から開始し、
    # n=0 の場合はラグ項を含まないモデルとして外部で計算する方がシンプルです。
    
    for k in range(1, n + 1):
        # dln_I を k 期分ずらした新しいラグ列を作成
        lag_col_name = f'I_dln_lag{k}'
        df[lag_col_name] = df["dln_I"].shift(k)
        X_lag_list.append(df[lag_col_name])

    # 3. データフレームの結合と欠損値の処理
    # 被説明変数(y), 基本のX_base, 全てのラグ変数を結合
    df_model = pd.concat([y.rename('dln_I')] + [X_base] + X_lag_list, axis=1).dropna()
    
    # 4. 推定の実行
    y_train = df_model['dln_I']
    # X_train は dln_I 以外の全て (dln_GDP, d_r, 全てのラグ)
    X_train = df_model.drop('dln_I', axis=1)
    
    model = sm.OLS(y_train, sm.add_constant(X_train)).fit()
    
    return model.aic

# =========================================================
# 実行例: 最小AICを探索するループ
# =========================================================
max_lag_to_test = 8  # 年次データなので、今回は最大8年分まで試してみましょう
aic_results = {}

print("--- 投資関数 (I) の最適ラグ探索 (変化率モデル) ---")

# 1. ラグなしモデル (n=0) を個別に計算
# n=0 のモデル: dln_I = i0 + i1*dln_GDP + i2*d_r
X_base_clean = df[['dln_GDP', 'd_r']].dropna()
Y_clean = df['dln_I'].loc[X_base_clean.index] # NaN除去後のインデックスに合わせる
model_lag0 = sm.OLS(Y_clean, sm.add_constant(X_base_clean)).fit()
aic_results[0] = model_lag0.aic
print(f"  ラグ 0: AIC = {model_lag0.aic:.4f}")


# 2. ラグ n=1 から max_lag までをループで計算
for n in range(1, max_lag_to_test + 1):
    current_aic = I_aic(n, df)
    aic_results[n] = current_aic
    print(f"  ラグ {n}: AIC = {current_aic:.4f}")

# 最小AICの特定
best_n = min(aic_results, key=aic_results.get)
best_aic = aic_results[best_n]

print("-" * 50)
print(f"✨ 最適ラグ次数: {best_n} (最小AIC: {best_aic:.4f})")
```

```python
df['dln_C'] = np.log(df['N_C']).diff()
df['dln_Y_minus_T'] = np.log(df['Y_minus_T']).diff()

def C_aic(n, df):
    """
    消費関数モデルにラグ次数 n を含めて推定し、AICを計算します。
    被説明変数: dln_C (消費変化率)
    説明変数: dln_Y_minus_T (可処分所得変化率), dln_C のラグ k=1..n
    """
    
    # 1. 被説明変数 (y) と基本の説明変数 (X_base) の定義
    y = df["dln_C"]
    # 消費関数に必要な基本の説明変数は dln_Y_minus_T の1つです
    X_base = df[["dln_Y_minus_T"]]
    
    # 2. ラグ変数の作成と結合
    X_lag_list = []
    
    for k in range(1, n + 1):
        # dln_C を k 期分ずらした新しいラグ列を作成
        lag_col_name = f'C_dln_lag{k}'
        df[lag_col_name] = df["dln_C"].shift(k)
        X_lag_list.append(df[lag_col_name])

    # 3. データフレームの結合と欠損値の処理
    df_model = pd.concat([y.rename('dln_C')] + [X_base] + X_lag_list, axis=1).dropna()
    
    # 4. 推定の実行
    y_train = df_model['dln_C']
    X_train = df_model.drop('dln_C', axis=1)
    
    # OLSを実行し、AICを返します
    model = sm.OLS(y_train, sm.add_constant(X_train)).fit()
    return model.aic

# =========================================================
# 実行例: 最小AICを探索するループ
# =========================================================
max_lag_to_test = 8  # 投資関数と同様に最大8ラグまで試す
aic_results = {}

print("--- 消費関数 (C) の最適ラグ探索 (変化率モデル) ---")

# ラグなしモデル (n=0) を個別に計算
X_base_clean = df[['dln_Y_minus_T']].dropna()
Y_clean = df['dln_C'].loc[X_base_clean.index]
model_lag0 = sm.OLS(Y_clean, sm.add_constant(X_base_clean)).fit()
aic_results[0] = model_lag0.aic
print(f"  ラグ 0: AIC = {model_lag0.aic:.4f}")


# ラグ n=1 から max_lag までをループで計算
for n in range(1, max_lag_to_test + 1):
    current_aic = C_aic(n, df)
    aic_results[n] = current_aic
    print(f"  ラグ {n}: AIC = {current_aic:.4f}")

# 最小AICの特定
best_n = min(aic_results, key=aic_results.get)
best_aic = aic_results[best_n]

print("-" * 50)
print(f"✨ 最適ラグ次数: {best_n} (最小AIC: {best_aic:.4f})")
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from linearmodels.iv import IV2SLS # 2SLS推定専用ライブラリ

# ----------------------------------------------------
# 【データ準備の仮定】
# df は既に以下の対数差分変数を含んでいると仮定します:
# dln_C, dln_Y_minus_T, dln_G, dln_C_lag1, dln_C_lag2, ...
# ----------------------------------------------------

# --- 消費関数 (C) のためのデータセット準備 ---
# 最適ラグは 1 で確定済み。

# 1. ターゲット変数を定義
Y = df['dln_C']  # 被説明変数

# 2. 内生変数、外生変数、操作変数を定義

# 内生な説明変数 (X_endo): 所得変化
X_endo = df['dln_Y_minus_T']

# 外生な説明変数 (X_exog): ラグ項 (AICによりラグ1が最適)
X_exog = df['C_dln_lag1']  # dln_C の1期ラグ

# 操作変数 (Z_inst): 
# X_exog (dln_C_lag1) は既に外生なので Z に含める。
# さらに、内生変数を識別するための IV として、政府支出のラグを使用
Z_inst = pd.concat([df['C_dln_lag1'],  # 外生変数 X_exog は Z に含める
                     df['dln_G'].shift(1)], # 政府支出 (dln_G) の1期ラグを IV として追加
                    axis=1)

# 3. データの結合と欠損値の処理
# IV2SLS は欠損値に弱いため、先に全ての変数を結合し、NaN行を除去します。
# dropna() でインデックスを同期させる。
data_for_2sls = pd.concat([Y, X_endo, Z_inst], axis=1).dropna() 

# 4. 2SLSモデルの実行 (linearmodels.iv を使用)
# 変数名を設定し直す
Y_train = data_for_2sls['dln_C']
X_endo_train = data_for_2sls['dln_Y_minus_T']
X_exog_train = data_for_2sls['dln_C_lag1'] # ラグ項は外生として扱う
Z_inst_train = data_for_2sls.drop(['dln_C', 'dln_Y_minus_T'], axis=1) # IV (ラグC + ラグG)

# 定数項 (切片) の追加
X_exog_train = sm.add_constant(X_exog_train)

# IV2SLSを実行
# 構文: IV2SLS(従属変数, 外生変数, 内生変数, 操作変数)
model_c_2sls = IV2SLS(Y_train, X_exog_train, X_endo_train, Z_inst_train).fit(
    cov_type='robust' # 頑健標準誤差を使用 (時系列データで推奨)
)

print("="*60)
print("【消費関数 (C) の 2SLS 推定結果】")
print(model_c_2sls)
print("="*60)
```
